```python
import pandas as pd
import numpy as np

# Load raw data
df_raw = pd.read_excel(
    r'C:\Users\Sansuwa\Desktop\data\raw\deaths_2024.xlsx',
    sheet_name='Table 1.2',
    skiprows=4,
    header=[0, 1]
)

# Save original file
df_original = df_raw.copy()

print(f"Raw data loaded: {df_raw.shape[0]} rows, {df_raw.shape[1]} columns")
print(df_raw.head(3))
```

```python
# Working on a copy so the original stays safe
df = df_raw.copy()

# Giving first two columns proper names
df.columns = ['Cause', 'Unit'] + list(df_raw.columns[2:])

# Reshaping to long format — years and sex become their own rows
df_long = df.melt(
    id_vars=['Cause', 'Unit'],
    var_name='Year_Sex',
    value_name='Deaths'
)

# The multi-level columns already have year and sex separated
# So I just extract them directly
df_long['Year'] = df_long['Year_Sex'].apply(lambda x: x[0])
df_long['Sex'] = df_long['Year_Sex'].apply(lambda x: x[1])
df_long = df_long.drop(columns=['Year_Sex', 'Unit'])

print("Data reshaped!")
print(df_long.head(20))
print(f"\nColumns: {df_long.columns.tolist()}")
```

```python
# Check both sexes are present
print(df_long['Sex'].unique())

# Show counts
print(df_long['Sex'].value_counts())
```
